### Series

Series is similar to array in Numpy. Unlike Numpy array, Series can have axis labels, meaning it can be indexed by a label, instead of just a number location. It also doesn't need to hold numeric data, it can hold any arbitrary Python Object.

list, Numpy array or dictionary can be converted into series

```python
import numpy as np
import pandas as pd
```
