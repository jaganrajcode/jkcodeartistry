### Series

Series is similar to array in Numpy. Unlike Numpy array, Series can have axis labels, meaning it can be indexed by a label, instead of just a number location. It also doesn't need to hold numeric data, it can hold any arbitrary Python Object.

list, Numpy array or dictionary can be converted into series

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a','b','c','d','e']
my_list = [10,20,30,40,50]
arr = np.array([10,20,30,40,50])
dict = {'a':10,'b':20,'c':30,'d':40,'e':50}
```

```python
pd.Series(data=my_list) # use list to create series
```

```python
pd.Series(my_list,labels)
```

```python
pd.Series(arr) # Series from array
```

```python
pd.Series(arr,labels)
```

```python
pd.Series(dict) # Series from dictionary
```

```python
# Pandas Series can hold variety of objects
pd.Series(data=labels)
```

```python
# can also hold functions
pd.Series(data=[print,len,sum])
```

```python
# using index 
```

```python
series1 = pd.Series([1,2,3,3],index=['dog','cat','pig','goat'])
series1
```

```python
series2 = pd.Series([4,3,2,1],index=['sheep','cat','dog','pig'])
series2
```

```python
series2['sheep']
```

```python
series1 + series2
```
